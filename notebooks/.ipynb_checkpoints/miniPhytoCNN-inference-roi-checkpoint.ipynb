## Load the trained model (weights + structure)

Model is trained on Resnet18 with 6 classes:
1. Ceratium furca
2. Prorocentrum graciles
3. Akashiwo
4. Eucampia
5. Laboea
6. Dinophysis

Each class is trained with 400 images.


## Load Model

```python
# Load checkpoint
checkpoint_path = "./model/resnet18-6class.pth"
checkpoint = torch.load(checkpoint_path, map_location=torch.device("cuda"),weights_only=False)  # or "cuda" if available


# Recreate the model architecture
model = models.resnet18(weights=None)


# Modify first conv layer if input was grayscale
if checkpoint.get("input_channels", 3) == 1:
    model.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)


# Modify final FC layer
num_features = model.fc.in_features
model.fc = nn.Linear(num_features, checkpoint["num_classes"])


# Load weights
model.load_state_dict(checkpoint["model_state_dict"])
model.eval()


# Restore class labels
class_to_idx = checkpoint["class_to_idx"]
idx_to_class = {v: k for k, v in class_to_idx.items()}


print("Model loaded and ready for inference")
```

```python
class IFCBFileDataset(Dataset):
    def __init__(self, fileset_bin, transform=None):
        """
        This class takes a FilesetBin objects (ifcb.open_raw(path)) and will handle the images and the transform.
        
        :param fileset_bin: a FilesetBin object (with .adc and .roi files)
        :param transform: optional torchvision transforms to apply
        """
        self.fs = fileset_bin
        self.transform = transform
        self.roi_indices = list(self.fs.images.keys())  # target numbers

    def __len__(self):
        return len(self.roi_indices)

    def __getitem__(self, idx):
        roi_num = self.roi_indices[idx]
        with self.fs:  # ensures the ROI file is open
            img = self.fs.images[roi_num]  # numpy 2D array

        # Normalize and convert to float32
        img = img.astype(np.float32) / 255.0
        img = torch.from_numpy(img).unsqueeze(0)  # (1, H, W)

        if self.transform:
            img = self.transform(img)

        return img, roi_num  # return the index for tracking if needed
```

```python
# Transforms for ResNet18 (ImageNet compatible)
# Very important to normalize to whatever the model weights were originally normalized to
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),
    transforms.Resize((224, 224)),
    transforms.Normalize((0.485,), (0.228,))
])

dataset = IFCBFileDataset(roi, transform=transform)
loader = torch.utils.data.DataLoader(dataset, batch_size=64)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Store predictions
all_preds = []
all_probs = []
all_roi_ids = []

model.eval()
with torch.no_grad():
    for imgs, roi_ids in loader:
        imgs = imgs.to(device)
        outputs = model(imgs)
        probs = F.softmax(outputs, dim=1).cpu().numpy()
        preds = outputs.argmax(dim=1).cpu().numpy()
        
        all_preds.extend(preds)
        all_probs.extend(probs)
        all_roi_ids.extend(roi_ids.numpy())

# Convert predictions to class names using idx_to_class
predicted_labels = [idx_to_class[p] for p in all_preds]

# Convert probabilities to DataFrame and rename columns
prob_df = pd.DataFrame(all_probs, columns=[idx_to_class[i] for i in range(len(idx_to_class))])

# Final DataFrame
results_df = pd.DataFrame({
    "roi_id": all_roi_ids,
    "predicted_class": predicted_labels
})
results_df = pd.concat([results_df, prob_df], axis=1)


# Show results
results_df.head()
```
